```python
import pandas as pd
```

```python
distritos = pd.read_csv('../../datasets/distritos.csv', sep='\t')
barrios = pd.read_csv('../../datasets/barrios.csv', sep='\t')
```

```python
distritos
```

```python
barrios
```

```python
parking = pd.read_parquet('../../datasets/tickets_parking.parquet')
```

```python
parking.head()
```

```python
print(parking.info)
print(parking.dtypes)
```

```python
parking['fecha_operacion'] = pd.to_datetime(parking['fecha_operacion'], format="%Y-%m-%d %H:%M:%S")
parking['fecha_inicio'] = pd.to_datetime(parking['fecha_inicio'], format="%Y-%m-%d %H:%M:%S")
parking['fecha_fin'] = pd.to_datetime(parking['fecha_fin'], format="%Y-%m-%d %H:%M:%S")
```

```python
print(parking.dtypes)
```

```python
# Merge distritos
p_distritos = pd.merge(parking, distritos)
p_distritos.head()
```

```python
# Check dataframe. It's the same num of rows than before merge
print(parking.info)
```

```python
p_distritos[['cod_distrito','distrito']]
```

```python
# Check dataframe distinct values of cod_distrito
print(p_distritos[['cod_distrito']].value_counts())

# Check dataframe distinct values of distrito
print(p_distritos[['distrito']].value_counts())
```

```python
# Check dataframe distinct values of cod_distrito + distrito 
print(p_distritos[['cod_distrito','distrito']].value_counts())

count_distritos = p_distritos[['cod_distrito','distrito']].value_counts().count()
print(f'total distinct values cod_distrito + distrito = {count_distritos}')

sum_distritos = p_distritos[['cod_distrito','distrito']].value_counts().sum()
print(f'total rows cod_distrito + distrito = {sum_distritos}')
```

```python
# Missing NaN values of distrito column. Clean NaN
p_distritos['distrito'] = p_distritos['distrito'].fillna('SIN INFORMAR')
p_distritos[['cod_distrito','distrito']]
```

```python
# Check dataframe distinct values of cod_distrito + distrito, after clean NaN
print(p_distritos[['cod_distrito','distrito']].value_counts())

count_distritos = p_distritos[['cod_distrito','distrito']].value_counts().count()
print(f'total distinct values cod_distrito + distrito = {count_distritos}')

sum_distritos = p_distritos[['cod_distrito','distrito']].value_counts().sum()
print(f'total rows cod_distrito + distrito = {sum_distritos}')
```

```python
# Merge barrios
p_barrios = pd.merge(p_distritos, barrios)
p_barrios.head()
```

```python
# Check dataframe distinct values of cod_distrito + cod_barrio + distrito + barrio
print(p_barrios[['cod_distrito','cod_barrio','distrito','barrio']].value_counts())

count_barrios = p_barrios[['cod_distrito','cod_barrio','distrito','barrio']].value_counts().count()
print(f'total distinct values cod_distrito + cod_barrio + distrito + barrio = {count_barrios}')

sum_barrios = p_barrios[['cod_distrito','cod_barrio','distrito','barrio']].value_counts().sum()
print(f'total rows cod_distrito + cod_barrio + distrito + barrio = {sum_barrios}')
```

```python
stats_barrios = p_barrios.groupby(['distrito','barrio'])[['minutos_tique', 'importe_tique']]\
                  .agg(['min', 'max', 'mean', 'median', 'std'])\
                  .sort_values(by=('importe_tique', 'median'), ascending=False).reset_index()
stats_barrios
```

```python
print(p_barrios['importe_tique'].min())
print(p_barrios['importe_tique'].max())
```

```python
df_min_max = pd.concat([p_barrios[(p_barrios['importe_tique'] == p_barrios['importe_tique'].min())],
                        p_barrios[(p_barrios['importe_tique'] == p_barrios['importe_tique'].max())]])
df_min_max
```

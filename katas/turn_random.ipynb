```python
import pandas as pd
import random
```

```python
students = pd.DataFrame(['Guillaume', 'Pedro', 'Javier', 'Jaime', 'Elvira', 'Halima', 'Ángel', 'Fran', 'Ana'], 
                        columns = ['name'])
students
```

```python
turn_list = []
```

```python
turn = random.randint(0, 8)
if turn not in turn_list:
    turn_list.append(turn)
else:
    print("repetido")
students.iloc[turn_list]
```

```python
if len(turn_list) < 9:
    turn = random.randint(0, 8)
    while turn in turn_list:
        turn = random.randint(0, 8)
    turn_list.append(turn)
students.iloc[turn_list]
```
